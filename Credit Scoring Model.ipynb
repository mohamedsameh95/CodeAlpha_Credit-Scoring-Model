## Importing Modules:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
```

## Loading the Dataset:

```python
import os
for dirname, _, filenames in os.walk("Bank"):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("bankloans.csv")
df.head()
```

```python
df = df.dropna()
```

```python
fig,ax = plt.subplots(figsize=(20,30))
sns.lineplot(x='age',y='income',data=df,ax=ax)
```

```python
df['default'].value_counts()
```

```python
x=df.drop(['default'],axis=1)
y=df['default']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
sc = StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

## Creating Model

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(xtrain,ytrain)
rfc.score(xtest,ytest)
```

```python
rfc2 = cross_val_score(estimator=rfc,X=xtrain,y=ytrain,cv=10)
rfc2.mean()
```

```python
sv = SVC()
sv.fit(xtrain,ytrain)
```

```python
sv.score(xtest,ytest)
```

```python
model = GridSearchCV(sv,{
    'C':[0.1,0.2,0.4,0.8,1.2,1.8,4.0,7.0],
    'gamma':[0.1,0.4,0.8,1.0,2.0,3.0],
    'kernel':['rbf','linear']
},scoring='accuracy',cv=10)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.best_params_
```

```python
model = SVC(C=0.1,gamma=0.1,kernel='linear')
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
lr.score(xtest,ytest)
```

```python
yp = lr.predict(xtest)
c= confusion_matrix(ytest,yp)
fig ,ax = plt.subplots(figsize=(20,10))
sns.heatmap(c,ax=ax)
```
